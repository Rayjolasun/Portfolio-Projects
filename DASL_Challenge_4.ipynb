## Task 1

**Remove duplicate values from the movies column**

```python
# check the number of duplicates in the MOVIES column

num_duplicates = df['MOVIES'].duplicated().sum()
print("Number of duplicate values in 'MOVIES' column:", num_duplicates)
```

```python
# List of duplicated MOVIES column

column_name = 'MOVIES'

# Find duplicates and count their occurrences

duplicates = df[column_name].duplicated(keep=False)
duplicate_counts = df[duplicates][column_name].value_counts()

# Print the list of duplicates and their occurrence counts

duplicate_counts
```

```python
# drop the duplicate values in MOVIES

df.drop_duplicates(subset='MOVIES', inplace=True)
```

```python
# To confirm if there are still duplicate values in MOVIES

num_duplicates = df.MOVIES.duplicated().sum()
num_duplicates
```

```python
df.info()
```

## Task 2

**Clean the year column to get the appropriate year**

```python
# Extract the year using regular expressions
df['YEAR'] = df['YEAR'].str.extract(r'(\d{4})')
```

```python
# confirm that the appropriate year has been extracted

df.sample(10)
```

```python
# Count the missing values in each column
missing_counts = df.isnull().sum()

# Print the count of missing values in each column
print(missing_counts)
```

```python
# Replace null values and non-date values in the "YEAR" column with a default value or a specific treatment

df['YEAR'].fillna(0, inplace=True)  # Replace null values with 0 
df['YEAR'] = pd.to_numeric(df['YEAR'], errors='coerce')  # Convert non-date values to NaN

df.info()
```

```python
df.sample(10)
```

## Task 3

**Get the first genre from the genre column**

```python
# Extract the genre column by splitting it by ',' allowing only the first Genre stated to remain

df['GENRE'] = df['GENRE'].str.split(',').str[0].str.strip()

df.sample(10)
```

```python
# Replace null values in the "GENRE" column with "Not Available"
df['GENRE'].fillna('Not Available', inplace=True)

df.sample(10)
```

```python
# Count the missing values in each column
missing_counts = df.isnull().sum()

# Print the count of missing values in each column
print(missing_counts)
```

## Task 4

**Clean the rating column**

```python
# Check the number of empty rows in the Rating column

has_nan = df['RATING'].isna().sum()
print("NaN values in 'RATING' column:", has_nan)
```

```python
# Clean the Rating column by filling the empty rows with 'Not Available' since the ratings are not available

df['RATING'].fillna('Not Available', inplace=True)

df.head()
```

## Task 5

**Get the director name and put it in a different column and also put the stars names in a separate column**

```python

# Extract the director names
df['Directors'] = df['STARS'].str.extract('Director[s]?:(.+?)(?=\s*\||Stars:|$)', flags=re.S, expand=False)
df['Directors'] = df['Directors'].str.replace('\n', '')

# Extract the stars names
stars1 = df['STARS'].str.extract('Star[s]?:(.+)', flags=re.S, expand=False)
stars2 = df['STARS'].str.extract('Stars:(.+)', flags=re.S, expand=False)

# Concatenate star names into a single column
df['Stars'] = stars1.fillna('').values.flatten() + stars2.fillna('').values.flatten()
df['Stars'] = df['Stars'].str.replace('\n', '')

# Remove leading/trailing whitespace from the extracted columns
df['Directors'] = df['Directors'].str.strip()
df['Stars'] = df['Stars'].str.strip()

# Fill NaN or null values or empty values with ''
df['Directors'] = df['Directors'].fillna("")
df['Stars'] = df['Stars'].fillna("")

# Insert the 'director' and 'stars' columns in the original position
original_position = df.columns.get_loc('STARS')
df.insert(original_position, 'Directors', df.pop('Directors'))
df.insert(original_position + 1, 'Stars', df.pop('Stars'))

# Drop the 'STARS' column
df = df.drop('STARS', axis=1)

df.sample(20)
```

## Task 6

**Clean the votes column**


```python
has_nan = df['VOTES'].isna().sum()
print("NaN values in 'VOTES' column:", has_nan)
```

```python
df['VOTES'] = df['VOTES'].replace('NaN', np.nan)
```

```python
df['VOTES'] = df['VOTES'].str.replace(',', '')  # Remove commas
df['VOTES'] = pd.to_numeric(df['VOTES'], errors='coerce').fillna(0).astype(int)  # Convert to whole numbers

df.head()
```

## Task 7

**Clean the runtime column**


```python
has_nan = df['RunTime'].isna().sum()
print("NaN values in 'RunTime' column:", has_nan)
```

```python
df['RunTime'].fillna('Not Available', inplace=True)

df.head()
```

## Task 8

**Clean the gross column**

```python
has_nan = df['Gross'].isna().sum()
print("NaN values in 'Gross' column:", has_nan)
```

```python
df['Gross'] = df['Gross'].str.replace(',', '')  # Remove commas
df['Gross'] = df['Gross'].str.replace('[\$,M]', '', regex=True).astype(float) * 1e6

median_gross = df['Gross'].median()
df['Gross'].fillna(median_gross, inplace=True)  # Fill NaN with median
df['Gross'] = df['Gross'].astype(int)  # Convert to whole numbers (integer data type)


df.head()
```

```python
# Convert column names to proper sentence case
df.columns = df.columns.str.title()

df.head()
```

```python

# Export the cleaned data to a CSV file
df.to_csv('cleaned_movies.csv', index=False)
```
